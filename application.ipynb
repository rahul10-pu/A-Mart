	public void validate(Franchise franchise) throws Exception
	{
		try{
			if(!isValidQuotedPrice(franchise.getQuotedPrice()))
				throw new Exception("Validator.INVALID_QUOTED_PRICE");
			if(!isValidContactNo(franchise.getContactNo()))
				throw new Exception("Validator.INVALID_CONTACTNO");
			if(!isValidDates(franchise.getRegDate(),franchise.getStartupDate()))
			 throw new Exception("Validator.INVALID_DATE");
			if(!isValidFranchiseName(franchise.getName()))
				 throw new Exception("Validator.INVALID_FRANCHISE_NAME");
			if(!isValidTaxNo(franchise.getTaxNo()))
				throw new Exception("Validator.INVALID_TAXNO");
			
		}
		catch (Exception e) {
			LogConfig.getLogger(this.getClass()).error(e.getMessage(), e);
			throw e;
		}
		
	}

	/*

	public Boolean isValidContactNo(String contactNo)
	{
		String regex="[7-9][0-9]{9}";
		
		String pattern3=contactNo.substring(0,1)+"{10}";
		if(contactNo.matches(regex)&&!contactNo.matches(pattern3)){
			return true;
		}
		else{
		return false;
	}
	}
	
	
	/*
	 * DO NOT CHANGE THE METHOD SIGNATURE 
	 * DO NOT DELETE/COMMENT METHOD 
	 * CHANGE return STATEMENT AS PER YOUR IMPLEMENTATION
	 */
	public Boolean isValidDates(LocalDate regDate, LocalDate startupDate)
	{
		LocalDate today=LocalDate.now();
		if(regDate.isAfter(today)){
			if((startupDate.isEqual(regDate))||((startupDate.isAfter(regDate))&&!(regDate.until(startupDate, ChronoUnit.MONTHS) <= 6))){
				return true;
			}else{
				return false;
				}
			}
		
		else{
			return false;
		}
		
	}
		
	

	
	/*
	 * DO NOT CHANGE THE METHOD SIGNATURE 
	 * DO NOT DELETE/COMMENT METHOD 
	 * CHANGE return STATEMENT AS PER YOUR IMPLEMENTATION
	 */
	public Boolean isValidFranchiseName(String name)
	{
		String regex="[A-Za-z]+( [A-Za-z]+{0,})";
		if(name.startsWith(" ")&& name.endsWith(" ")){
			return false;
		}
		else{
			return name.matches(regex);
		}
		
	}
	
	
	/*
	 * DO NOT CHANGE THE METHOD SIGNATURE 
	 * DO NOT DELETE/COMMENT METHOD 
	 * CHANGE return STATEMENT AS PER YOUR IMPLEMENTATION
	 */
	public Boolean isValidTaxNo(String taxNo)
	{
		String []s=taxNo.split("-");
		String s1="[0-9]{2}";
		String s2="[A-Z]{8}";
		String s3="[0-9]{1,}";
		String s4="[I]";
		String s5="[1|5]";
		if((s[0].matches(s1))&&(s[1].matches(s2))&&(s[2].matches(s3))&&(s[3].matches(s4))&&(s[4].matches(s5))){
			return true;
		}
		else{
		return false;
	}
	}
